Dada a imagem anexa, use apenas filtros de convolução 3x3 e thresholding para produzir uma imagem apenas com o contorno de um triângulo.

```python
import cv2

import numpy as np
import seaborn as sns

from math import sqrt
from random import choice

from PIL import Image
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
f = 'warning-triangle.jpg'
x = cv2.imread(f)
x.shape
```

```python
__show = lambda i: cv2.cvtColor(i, cv2.COLOR_BGR2RGB)
```

```python
plt.title(f'{f} {x.shape}')
plt.imshow(__show(x))
plt.show()
```

```python
fbordas = np.array([
    [-1, 1, 1],
    [-1, 0, 1],
    [-1, 1, 1]])

conv = lambda i,f: cv2.filter2D(i, -1, f)
tresh = lambda i: cv2.threshold(i, 50, 255, cv2.THRESH_BINARY)[-1]

plt.figure(figsize=(12,8))

__ = (
    conv(x, fbordas),
    tresh(conv(x, fbordas))
 )

for i,_ in enumerate(__):
    plt.subplot(len(__),2,i+1)
    plt.imshow(__show(_))
    plt.xticks([])
    plt.yticks([])

plt.show()
```
