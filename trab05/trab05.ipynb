```python
import cv2

import numpy as np
import seaborn as sns

from math import sqrt
from random import choice

from PIL import Image
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
f1, f2 = 'img01.jpg', 'img02.jpg'

x1, x2 = cv2.imread(f1), cv2.imread(f2)
```

```python
plt.subplot(1,2,1)
plt.title(str(x1.shape))
plt.imshow(x1)

plt.subplot(1,2,2)
plt.title(str(x2.shape))
plt.imshow(x2)

plt.show()
```

```python
def equalizacao(imagem):
    imagem = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)
    imagem = cv2.equalizeHist(imagem)
    return imagem
```

```python
y1eq, y2eq = equalizacao(x1), equalizacao(x2)

y1eq.shape, y2eq.shape
```

```python
plt.figure(figsize=(8,8))

plt.subplot(2,2,1)
_ = x1
__ = _.shape
plt.title(f'original {__}')
plt.imshow(_)
plt.xticks([])
plt.yticks([])

plt.subplot(2,2,2)
_ = x2
__ = _.shape
plt.title(f'original {__}')
plt.imshow(_)
plt.xticks([])
plt.yticks([])

plt.subplot(2,2,3)
_ = y1eq
__ = _.shape
plt.title(f'equalizacao {__}')
plt.imshow(_, cmap='gray')
plt.xticks([])
plt.yticks([])

plt.subplot(2,2,4)
_ = y2eq
__ = _.shape
plt.title(f'equalizacao {__}')
plt.imshow(_, cmap='gray')
plt.xticks([])
plt.yticks([])

plt.show()
```

___

```python
def redimensionar(imagem, horizontal=1.0, vertical=1.0):
    altura_original, largura_original = imagem.shape[0], imagem.shape[1], 
    nova_altura = int(altura_original * horizontal)
    nova_largura = int(largura_original * vertical)
    imagem_redimensionada = cv2.resize(imagem, (nova_largura, nova_altura))
    return imagem_redimensionada
```

```python
y1rd, y2rd = redimensionar(x1, 1.23, 1.0), redimensionar(x2, 0.5, 2.0)

y1rd.shape, y2rd.shape
```

```python
plt.figure(figsize=(8,8))

plt.subplot(2,2,1)
_ = x1
__ = _.shape
plt.title(f'original {__}')
plt.imshow(_)
plt.xticks([])
plt.yticks([])

plt.subplot(2,2,2)
_ = x2
__ = _.shape
plt.title(f'original {__}')
plt.imshow(_)
plt.xticks([])
plt.yticks([])

plt.subplot(2,2,3)
_ = y1rd
__ = _.shape
plt.title(f'resize {__}')
plt.imshow(_)
plt.xticks([])
plt.yticks([])

plt.subplot(2,2,4)
_ = y2rd
__ = _.shape
plt.title(f'resize {__}')
plt.imshow(_)
plt.xticks([])
plt.yticks([])

plt.show()
```
